**P2.** Dada la lista `numeros = [10, 25, 3, 47, 8, 15]`:  
- Imprime el primer y el último elemento.  
- Imprime los tres primeros elementos usando slicing.  
- Agrega el número `99` al final de la lista.

```python
numeros = [10, 25, 3, 47, 8, 15]

# Tu respuesta aquí
# Primer y último elemento
print(f"El primer elemento es {numeros[0]} y el último elemento es {numeros[-1]}")


# Tres primeros elementos (slicing)
print(f"Los tres primeros elementos son: {numeros[:3]}" )

# Agregar 99 al final
numeros.append(99)
print(numeros)
```

**P3.** Dado el diccionario `estudiante` con las llaves `nombre`, `carrera` y `promedio`,  
escribe un `if` que imprima `"Aprobado"` si el promedio es mayor o igual a 6, y `"Reprobado"` si no.

```python
estudiante = {"nombre": "Carlos", "carrera": "Sistemas", "promedio": 7.5}

# Tu respuesta aquí
if estudiante["promedio"] >= 6:
    estado = "Aprobado"
else:
    estado = "Reprobado"

print(f"Nombre: {estudiante['nombre']}, Carrera: {estudiante['carrera']}, Promedio: {estudiante['promedio']} → {estado}")
```

**P4.** Crea una función llamada `calcular_area` que reciba `base` y `altura` y devuelva el área de un triángulo (`base * altura / 2`).  
Llama la función con `base=5` y `altura=8` e imprime el resultado.

```python
# Tu respuesta aquí
def calcular_area(base, altura):
    return (base * altura) / 2

area = calcular_area(5, 8)
print(area)
```

**P5.** Usa una **list comprehension** para crear una lista con los cuadrados de los números del 1 al 10.

**P6.** Crea los siguientes arreglos e imprime su `shape` y `dtype`:  
- Un arreglo con los valores `[1.5, 2.5, 3.5, 4.5]`  
- Una matriz de ceros de 3 filas × 4 columnas  
- Un arreglo con 6 valores igualmente espaciados entre 0 y 1 (usando `linspace`)

```python
# Tu respuesta aquí
# 1. Arreglo con valores dados
arr1 = np.array([1.5, 2.5, 3.5, 4.5])
print("Arreglo 1:", arr1)
print("Shape:", arr1.shape)
print("Dtype:", arr1.dtype)

# 2. Matriz de ceros 3x4
arr2 = np.zeros((3, 4))
print("\nMatriz de ceros:\n", arr2)
print("Shape:", arr2.shape)
print("Dtype:", arr2.dtype)

# 3. Arreglo con linspace
arr3 = np.linspace(0, 1, 6)
print("\nArreglo linspace:", arr3)
print("Shape:", arr3.shape)
print("Dtype:", arr3.dtype)
```

**P7.** Dado el arreglo `a = np.arange(1, 13)`, usa `reshape` para convertirlo en una matriz de 3 × 4.  
Luego imprime el valor máximo, el valor mínimo y la posición (índice) del máximo.

```python
a = np.arange(1, 13)

# Tu respuesta aquí
# Convertir a matriz 3x4
matriz = a.reshape(3, 4)
print(matriz)

# Máximo, mínimo e índice del máximo
print("Máximo:", matriz.max())
print("Mínimo:", matriz.min())
print("Índice del máximo:", matriz.argmax())
```

**P8.** Crea dos arreglos de 5 elementos con `np.random.randint` (valores entre 1 y 20).  
Súmalos, réstalos y multiplícalos elemento por elemento. Imprime cada resultado.

```python
# Tu respuesta aquí
# Crear dos arreglos aleatorios
a = np.random.randint(1, 21, 5)
b = np.random.randint(1, 21, 5)

print("Arreglo a:", a)
print("Arreglo b:", b)

# Operaciones elemento a elemento
print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicación:", a * b)
```

**P9.** Dado el arreglo `notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])`,  
usa **indexación booleana** para obtener solo las notas mayores o iguales a 6. Imprime el resultado.

```python
notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])

# Tu respuesta aquí
print(f"Notas mayores o iguales a 6: {notas[notas >= 6]}")
```
