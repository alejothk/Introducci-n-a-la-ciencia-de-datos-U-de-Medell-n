## Parte 1 — Exploración y selección

**P1.** Imprime los nombres de todas las columnas y el número total de filas y columnas del DataFrame.

```python
# Tu respuesta aquí
print("Nombres de las columnas:", df.columns.tolist())
print("Número total de filas:", df.shape[0])
print("Número total de columnas:", df.shape[1])
```

**P2.** Usa `.loc` para seleccionar las filas 10 a 15 (inclusive) y solo las columnas `Name`, `Age` y `Survived`.

## Parte 2 — Valores nulos y limpieza

**P3.** Cuenta cuántos valores nulos hay en cada columna. ¿Cuáles columnas tienen nulos?

```python
# Tu respuesta aquí
df.isnull().sum()
```

**Respuesta:** Las columnas con valores nulos son: `Age`, `Cabin`, `Embarked`.

**P4.** Rellena los valores nulos de la columna `Age` con la **mediana** de esa columna.  
Verifica que ya no haya nulos en `Age`.

```python
# Tu respuesta aquí
age_median = df['Age'].median()
df['Age'].fillna(age_median, inplace=True)
print("Valores nulos en 'Age' después de rellenar:")
print(df['Age'].isnull().sum())
```

---
## Parte 3 — Nuevas columnas y agrupaciones

**P5.** Crea una nueva columna llamada `FamilySize` que sea la suma de `SibSp` + `Parch` + 1 (el pasajero mismo).

```python

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

**P6.** Usando `groupby`, calcula la **tasa de supervivencia promedio** (columna `Survived`) por clase (`Pclass`).  
¿Qué clase tuvo mayor probabilidad de sobrevivir?

```python
# Tu respuesta aquí
survival_rate_by_pclass = df.groupby('Pclass')['Survived'].mean()
print(survival_rate_by_pclass)
```

**Respuesta:** La clase con mayor supervivencia fue: `Clase 1`

**P7.** Calcula la edad promedio y la tarifa (`Fare`) promedio, agrupando por `Sex` y `Pclass`.  
Usa `.groupby()` con `.agg()`.

```python
# Tu respuesta aquí
average_age_fare_by_sex_pclass = df.groupby(['Sex', 'Pclass']).agg({'Age': 'mean', 'Fare': 'mean'})
print(average_age_fare_by_sex_pclass)
```

## Parte 4 — Ordenamiento y duplicados

**P8.** Ordena el DataFrame por `Fare` de mayor a menor y muestra las primeras 5 filas.  
¿Cuánto pagó el pasajero más caro?

```python
# Tu respuesta aquí
df_sorted_fare = df.sort_values(by='Fare', ascending=False)
display(df_sorted_fare.head(5))
richest_fare = df_sorted_fare.iloc[0]['Fare']
print(f"El pasajero más caro pagó: {richest_fare:.2f}")
```

**P9.** Verifica si hay filas duplicadas en el DataFrame. ¿Cuántas hay?

```python
# Tu respuesta aquí
num_duplicates = df.duplicated().sum()
print(f"Número de filas duplicadas: {num_duplicates}")
```
